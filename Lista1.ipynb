## Resolução:

O que temos de fazer é resolver a equação quadrática, portanto 

\begin{align}
a  &= 1\\
b  &= -\frac{2GM}{v_1 l_1} v_2\\ 
c  &= \frac{2GM}{l_1} - v_1 ^2\\
\end{align}
portanto basta fazer bhaskara com o python e encontrar a raiz, esta uma vez encontrada aplicam-se as formulas e obtêm-se os valores pedidos

```python
import numpy as np
import math
import matplotlib.pyplot as pl

G = 6.6738*10**(-11) #m^-3 kg^-1 s^-2
M = 1.9891*10**(30) #kg


v1 =30300 # m/s #float("insira v1:" input())
l1 =147000000000 #m #float("insira l1:" input())

def resolver_bhaskara(G,M,v1,l1):
    a = 1
    b = -((2*G*M)/(v1*l1))
    c =-v1**2+2*G*M/l1
    raiz = b**2 - 4*a*c
    if raiz < 0:
        return("A raiz é negativa")
    v2 = (-b - np.sqrt(raiz)) / (2*a)
    return v2
#print("Este é o valor da velocidade no afelio:",resolver_bhaskara(G,M,v1,l1))
v2 = resolver_bhaskara(G, M, v1, l1)

def raio2(v1,l1,v2):
    l2 = (v1*l1)/v2
    return l2

l2 = raio2(v1,l1,v2)


a = (l1+l2)/2
b = np.sqrt(l1*l2)

def periodo_orbital(a,b,l1,v1):
    T = (2*np.pi*a*b)/(l1*v1)
    return T
print("Em anos o periodo é:",periodo_orbital(a, b, l1, v1)/(60*60*24*365.25))
#Seria 1 se colocassemos 365,25, mas como estamos contabilizando o valor de um ano que é 365 dias
#Para fazer ao halley basta substituir os valores de v1 e l1, pode tambem deixar aberto como input sendo v1 = float("insira:"input()) e l1 = float("insira:"input())
```
